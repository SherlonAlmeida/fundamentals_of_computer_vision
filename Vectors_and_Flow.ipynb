# Calculando Flows

References: [1](https://matplotlib.org/3.1.1/gallery/images_contours_and_fields/quiver_simple_demo.html#sphx-glr-gallery-images-contours-and-fields-quiver-simple-demo-py) [2](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.quiver.html) [3](https://www.geeksforgeeks.org/quiver-plot-in-matplotlib/)

Basta definir um array de coordenadas (X,Y) e um array de direcoes (U,V).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_flow(X, Y, U, V):
    fig, ax = plt.subplots(figsize=(5,3))
    q = ax.quiver(X, Y, U, V)
    ax.invert_yaxis()
    ax.xaxis.tick_top()
    fig.tight_layout()
    plt.show()
```

## Plotando Flow em 2 Dimensoes (Como o RAFT)

Ou seja, cada pixel pode ir para qualquer lugar na proxima imagem.

```python
m,n = 4, 4
X = np.arange(0, m, 1)
Y = np.arange(0, n, 1)
U, V = np.meshgrid(X, Y)
plot_flow(X, Y, U, V)
```

```python
X, Y #Posicoes
```

```python
U, V #Direcoes
```

## Plotando Flow em 1 Dimensao (Como o RAFT-Stereo)

Ou seja, cada pixel pode ir para qualquer lugar NA MESMA EPIPOLAR LINE na proxima imagem. Isto significa que o vetor V deve ser de zeros.

```python
m,n = 4, 4
X = np.arange(0, m, 1)
Y = np.arange(0, n, 1)
U, _ = np.meshgrid(X, Y)
V = np.zeros((m,n))
plot_flow(X, Y, U, V)
```

```python
X, Y #Posicoes
```

```python
U, V #Direcoes
```

# Conclusao

Ao representar cada posicao da imagem (pixels) como um vetor (X,Y,U,V) indicando sua correspondencia na imagem seguinte, o que temos na verdade é que a magnitude do vetor representa a disparidade (em rectified Stereo-Vision).

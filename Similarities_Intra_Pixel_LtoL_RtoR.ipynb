```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
height, width = 4, 4
img = np.random.rand(height, width) #np.arange(height*width).reshape(height, width)
img_shift = np.arange(height*width).reshape(height, width)
```

```python
e, l = 0, 0 # Shift (Center)
fig, ax = plt.subplots(1, 1, figsize=(3, 2))
sns.heatmap(img, annot=True, cmap="Blues", ax=ax)
ax.set_title("Image")
plt.tight_layout()
plt.show()
```

# Entendendo Matrizes e Deslocamentos

```python
img_shift = np.roll(img, 1, axis=0) #Desloca para baixo
img_shift[0] = 0
img_shift
```

```python
img_shift = np.roll(img, -1, axis=0) #Desloca para cima
img_shift[height - 1] = 0
img_shift
```

```python
img_shift = np.roll(img, 1, axis=1) #Desloca para direita
img_shift[:,0] = 0
img_shift
```

```python
img_shift = np.roll(img, -1, axis=1) #Desloca para esquerda
img_shift[:,width-1] = 0
img_shift
```

```python
img_shift = np.roll(img, (-1,-1), axis=(0, 1)) #Desloca para cima-esquerda
img_shift[:,width-1] = 0
img_shift[height-1, :] = 0
img_shift
```

```python
img_shift = np.roll(img, (-1,1), axis=(0, 1)) #Desloca para cima-direita
img_shift[:, 0] = 0
img_shift[height-1, :] = 0
img_shift
```

```python
img_shift = np.roll(img, (1,-1), axis=(0, 1)) #Desloca para baixo-esquerda
img_shift[0, :] = 0
img_shift[:, width-1] = 0
img_shift
```

```python
img_shift = np.roll(img, (1,1), axis=(0, 1)) #Desloca para baixo-direita
img_shift[0, :] = 0
img_shift[:, 0] = 0
img_shift
```

#Automatizando o Deslocamento

```python
idx = [ [ 0, 0], #Centro
        [-1, 0], #Cima
        [ 1, 0], #Baixo
        [ 0,-1], #Esquerda
        [ 0, 1], #Direita
        [-1,-1], #Cima-Esquerda
        [-1, 1], #Cima-Direita
        [ 1,-1], #Baixo-Esquerda
        [ 1, 1]] #Baixo-Direita
```

```python
fig, ax = plt.subplots(3, 3, figsize=(8, 7))
for e, l in idx:
    img_shift = np.roll(img, (e, l), axis=(0, 1)) #Desloca para baixo

    if e == -1:
        img_shift[height-1, :] = 0
    if e == 1:
        img_shift[0, :] = 0
    if l == -1:
        img_shift[:, width-1] = 0
    if l == 1:
        img_shift[:, 0] = 0

    sns.heatmap(img_shift, annot=True, cmap="Blues", ax=ax[e+1][l+1])
    ax[e+1][l+1].set_title(f"Shift (e, l) = ({e},{l})")
plt.tight_layout()
plt.show()
```

# Calculando Correlaes

Uma vez que a similaridade dos dados pode ser mensurada atravs de uma multiplicao de matrizes, podemos obter a similaridade de cada pixel de uma imagem com os seus respectivos vizinhos (8-neighborhood) atravs do deslocamento da matriz.

```python
fig, ax = plt.subplots(3, 3, figsize=(8, 7))
for e, l in idx:
    img_shift = np.roll(img, (e, l), axis=(0, 1)) #Desloca para baixo

    if e == -1:
        img_shift[height-1, :] = 0
    if e == 1:
        img_shift[0, :] = 0
    if l == -1:
        img_shift[:, width-1] = 0
    if l == 1:
        img_shift[:, 0] = 0

    feat1 = np.expand_dims(img, axis=2)
    feat2 = np.expand_dims(img_shift, axis=2)

    similarities = np.round(np.multiply(feat1, feat2).squeeze(), 2)

    sns.heatmap(similarities, annot=True, cmap="Blues", ax=ax[e+1][l+1])
    ax[e+1][l+1].set_title(f"Similarities ({0},{0})->({e},{l})")
plt.tight_layout()
plt.show()
```

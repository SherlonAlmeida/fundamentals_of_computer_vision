```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gaussian function
def gaussian(x, mu, sigma):
    return (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
```

```python
# Plotting
def plot_figure(x, y):
    plt.figure(figsize=(4, 3))
    plt.plot(x, y, label='Gaussian function')
    plt.title('Gaussian Function')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Parameters for the Gaussian function
mu = 0       # mean
sigma = 1    # standard deviation

# Generating x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Generating y values
y = gaussian(x, mu, sigma)

plot_figure(x, y)
```

```python
# Parameters for the Gaussian function
mu = 10       # mean
sigma = 1    # standard deviation

# Generating x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Generating y values
y = gaussian(x, mu, sigma)

plot_figure(x, y)
```

## Definindo uma funo gaussiana centrada na mdia definida por **mu**

```python
# Parameters
n_values = 64 # Total number of values

# Generating x values
x = np.arange(n_values)
```

```python
x
```

```python
# Parameters for the Gaussian function
mu = 10        # mean
sigma = 5     # standard deviation

# Generating y values
y1 = gaussian(x, mu, sigma)

plot_figure(x, y1)
```

```python
# Parameters for the Gaussian function
mu = 32        # mean
sigma = 5     # standard deviation

# Generating y values
y2 = gaussian(x, mu, sigma)

plot_figure(x, y2)
```

```python
# Parameters for the Gaussian function
mu = 48        # mean
sigma = 5     # standard deviation

# Generating y values
y3 = gaussian(x, mu, sigma)

plot_figure(x, y3)
```

## Ponderando um sinal atravs de uma funo gaussiana

```python
sinal = np.ones(n_values) * 255
sinal
```

```python
y1_norm  = y1 / np.max(y1)  #Normaliza a funo gaussiana entre [0,1]
y1_gauss = sinal * y1_norm  #Aplica a ponderao com a funo gaussiana
y1_inv   = 255 - y1_gauss   #Inverte o sinal, para que o ponto mnimo seja a recompensa (no o mximo)
plot_figure(x, y1_inv)      #Apresenta a figura
```

```python
y2_norm  = y2 / np.max(y1)  #Normaliza a funo gaussiana entre [0,1]
y2_gauss = sinal * y2_norm  #Aplica a ponderao com a funo gaussiana
y2_inv   = 255 - y2_gauss   #Inverte o sinal, para que o ponto mnimo seja a recompensa (no o mximo)
plot_figure(x, y2_inv)      #Apresenta a figura
```

```python
y3_norm  = y3 / np.max(y3)  #Normaliza a funo gaussiana entre [0,1]
y3_gauss = sinal * y3_norm  #Aplica a ponderao com a funo gaussiana
y3_inv   = 255 - y3_gauss   #Inverte o sinal, para que o ponto mnimo seja a recompensa (no o mximo)
plot_figure(x, y3_inv)      #Apresenta a figura
```

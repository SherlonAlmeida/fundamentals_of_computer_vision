# Visualizing Derivatives

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plotting
def plot_figure(x, y, px, py, pxs, pys):
    plt.figure(figsize=(4, 3))
    plt.plot(x, y, label='Function') #Function
    plt.plot(px, py, 'o') #Point for the derivative
    plt.plot(pxs, pys) #Tangent Line at P
    plt.title('Function')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Obtaining Slope (m) where $m = \lim_{h\to0} \frac{f(x + h)-f(x)}{(x+h) - x}$.


```python
# Choose a Function
def f_x(x):
    return 1/(1+np.exp(-x)) # Sigmoid

#Derivative (Using limits)
def df_x(f, x, h):
    return (f(x + h) - f(x)) / ((x + h) - x)
```

```python
# Generating x values
x = np.linspace(-10, 10, 100) #Start, End, Total Discrete Points

# Generating y values
y = f_x(x)

# Obtaining Slope (Derivative) at P(x,y)
idx = 75
px, py = x[idx], y[idx]
m = df_x(f_x, px, h=0.00001)
print(f"px={px}, py={py}, m={m}")

#Obtaining 2 points from Line Equation at P(x,y)
interval = 2
px1 = px - interval #Left Point
py1 = -(m*(px-px1)-py)

px2 = px + interval #Right Point
py2 = -(m*(px-px2)-py)

plot_figure(x, y, px, py, [px1,px2], [py1,py2])
```
